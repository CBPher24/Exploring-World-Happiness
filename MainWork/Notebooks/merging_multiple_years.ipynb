```python
#grabbing data needed for merge
y2022_df_e = y2022_df.loc[: , ["RANK", "Country", "Happiness score"]]
y2021_df_e = y2021_df.loc[: , ["Country name", "Ladder score"]]
y2020_df_e = y2020_df.loc[: , ["Country name", "Ladder score"]]
y2019_df_e = y2019_df.loc[: , ["Overall rank", "Country or region", "Score"]]
y2018_df_e = y2018_df.loc[: , ["Overall rank", "Country or region", "Score"]]
y2017_df_e = y2017_df.loc[: , ["Happiness.Rank", "Country", "Happiness.Score"]]
y2016_df_e = y2016_df.loc[: , ["Happiness Rank", "Country", "Happiness Score"]]
y2015_df_e = y2015_df.loc[: , ["Happiness Rank", "Country", "Happiness Score"]]
```

```python
y2022_ren = y2022_df_e.rename(columns={"Country" : "Country",
                                              "RANK":"2022 Rank", "Happiness score":"2022 Happiness Score"})
y2021_ren = y2021_df_e.rename(columns={"Country name" : "Country",
                                              "2021 Rank":"2021 Rank", "Ladder score":"2021 Happiness Score"})
y2020_ren = y2020_df_e.rename(columns={"Country name" : "Country",
                                              "2020 Rank":"2020 Rank", "Ladder score":"2020 Happiness Score"})
y2019_ren = y2019_df_e.rename(columns={"Country or region" : "Country",
                                              "Overall rank":"2019 Rank", "Score":"2019 Happiness Score"})
y2018_ren = y2018_df_e.rename(columns={"Country or region" : "Country",
                                              "Overall rank":"2018 Rank", "Score":"2018 Happiness Score"})
y2017_ren = y2017_df_e.rename(columns={"Country" : "Country",
                                              "Happiness.Rank":"2017 Rank", "Happiness.Score":"2017 Happiness Score"})
y2016_ren = y2016_df_e.rename(columns={"Country" : "Country",
                                              "Happiness Rank":"2016 Rank", "Happiness Score":"2016 Happiness Score"})
y2015_ren = y2015_df_e.rename(columns={"Country" : "Country",
                                              "Happiness Rank":"2015 Rank", "Happiness Score":"2015 Happiness Score"})
```

```python
sub_comb_data =  pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(y2022_ren,
                                                                 y2021_ren, on="Country", how="left")
                                                                ,y2020_ren, on="Country", how="left")
                                                                ,y2019_ren, on="Country", how="left")
                                                                ,y2018_ren, on="Country", how="left")
                                                                ,y2017_ren, on="Country", how="left")
                                                                ,y2016_ren, on="Country", how="left")
                                                                ,y2015_ren, on="Country", how="left").dropna()
sub_comb_data.loc[(sub_comb_data.Country == 'United States'),'Country']='United States of America'
yearly_data = sub_comb_data.set_index("Country")                                                            
```

```python
yearly_data = yearly_data[["2022 Rank","2022 Happiness Score","2021 Rank","2021 Happiness Score","2020 Rank",
                          "2020 Happiness Score", "2019 Rank", "2019 Happiness Score", "2018 Rank", 
                          "2018 Happiness Score", "2017 Rank", "2017 Happiness Score", "2016 Rank", 
                          "2016 Happiness Score", "2015 Rank", "2015 Happiness Score"]]
yearly_data = yearly_data.round({"2022 Happiness Score":2, "2021 Happiness Score":2, "2020 Happiness Score":2, 
                                 "2019 Happiness Score":2, "2018 Happiness Score":2, "2017 Happiness Score":2, 
                                 "2016 Happiness Score":2, "2015 Happiness Score":2})
yearly_data = yearly_data.astype({"2022 Rank": "int32", "2021 Rank": "int32", "2020 Rank": "int32",
                                  "2019 Rank": "int32", "2018 Rank": "int32", "2017 Rank": "int32",
                                  "2016 Rank": "int32", "2015 Rank": "int32",})
            
yearly_data
```
